```python
import smtplib
import pandas as pd
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
```

```python
# Load client emails
df = pd.read_csv("client_list.csv")
email_list = df['Email'].tolist()
```

```python
# Your email credentials
your_email = "danielamponsahkwakye@gmail.com"
your_password = "haoo jsfb digo puky"  # Use App Password if 2FA is enabled

# Email content
subject = "Quick idea to help you see your numbers differently"
body = """\
Hello!,

I know you're incredibly busy and probably get a ton of messages, so I’ll keep this super short.

I run a data analytics consultancy where we help businesses like yours uncover hidden insights in your operations — through dashboards, automation, and predictive modeling.

If I could show you a way to turn your everyday business data into decisions that save time, reduce stockouts, or boost profit — would you be open to a 15-minute call?

A quick question: have you ever wondered what your sales or inventory would look like if you could see tomorrow’s numbers today?

I’m happy to send over a quick dashboard sample, or even do a free mini-audit just to show you what’s possible.

When would be a good time for you to explore this — even briefly?

Totally understand if now’s not the right time. Even a one-line reply would completely make my day.

All the best,  
Daniel Amponsah Kwakye  
Data Analytics Consultant | Power BI | Predictive Insights  
www.linkedin.com/in/daniel-a-kwakye
"""

# Send emails
with smtplib.SMTP('smtp.gmail.com', 587) as server:
    server.starttls()
    server.login(your_email, your_password)

    for recipient in email_list:
        msg = MIMEMultipart()
        msg['From'] = your_email
        msg['To'] = recipient
        msg['Subject'] = subject

        msg.attach(MIMEText(body, 'plain'))

        try:
            server.sendmail(your_email, recipient, msg.as_string())
            print(f"Email sent to {recipient}")
        except Exception as e:
            print(f"Failed to send email to {recipient}: {e}")
```
